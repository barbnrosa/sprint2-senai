```python
from sklearn.preprocessing import LabelEncoder
s = (df_pf.dtypes == 'object')
object_cols = list(s[s].index)
LE=LabelEncoder()
for i in object_cols:
    df_pf[i]=df_pf[[i]].apply(LE.fit_transform)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_pf
```

```python
std=StandardScaler()
df_scaled = df_pf.copy()

colunas = ['Linha de Ação', 'Produto GPOM', 'Classe de Serviço',
                    'Cidade', 'Microrregião', 'Estado', 'Área de Atuação',
                    'Área de Atuação e Produto GPOM', 'Idade', 'Tempo de Contrato']

df_scaled[colunas] = std.fit_transform(df_scaled[colunas])
```

## 2. TREINAMENTOS

## 2.1 Treinamento da  Árvore de Decisão:

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

def treinar_e_avaliar_produto(nome_produto, df_scaled):
    # Dividir os dados em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(
        df_scaled[['Idade', 'Gênero', 'Microrregião']],
        df_scaled[f'Target{nome_produto}'],
        test_size=0.3,
        random_state=42
    )

    # Treinar árvore de decisão
    modelo = DecisionTreeClassifier()
    modelo.fit(X_train, y_train)

    
    y_pred = modelo.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Acurácia para {nome_produto}: {accuracy}')
    print(classification_report(y_test,y_pred))
    return modelo, accuracy

modelo_produtoA, accuracy_produtoA = treinar_e_avaliar_produto('A', df_scaled)

modelo_produtoB, accuracy_produtoB = treinar_e_avaliar_produto('B', df_scaled)

modelo_produtoC, accuracy_produtoC = treinar_e_avaliar_produto('C', df_scaled)

modelo_produtoD, accuracy_produtoC = treinar_e_avaliar_produto('D', df_scaled)
```

## Identificação do Público-Alvo e Atribuição de Segmentos

```python
def assign_segments(df_scaled, model, product_name, features=['Idade', 'Gênero', 'Microrregião']):

    # Fazer previsões usando o modelo
    predictions = model.predict(df_scaled[features])

    # Adicionar coluna de segmento ao DataFrame
    df_scaled[f'Segmento_{product_name}'] = predictions.astype(int)

# Chamar a função para cada modelo e produto
assign_segments(df_scaled, modelo_produtoA, 'ProdutoA')
assign_segments(df_scaled, modelo_produtoB, 'ProdutoB')
assign_segments(df_scaled, modelo_produtoC, 'ProdutoC')
assign_segments(df_scaled, modelo_produtoD, 'ProdutoD')
```

## Sistema de Recomendação

```python
def sistema_recomendacao(cliente):
    if cliente['Segmento_ProdutoA'] == 1:
        return "Recomendar ProdutoA"
    elif cliente['Segmento_ProdutoB'] == 1:
        return "Recomendar ProdutoB" 
    elif cliente['Segmento_ProdutoC'] == 1:
        return "Recomendar ProdutoC"
    elif cliente['Segmento_ProdutoD'] == 1:
        return "Recomendar ProdutoD"
    else:
        return "Sem recomendação neste momento"
```

```python
# Printa Somente os clientes que tem alguma recomendação
for cliente_id in df_scaled['ID Cliente'].unique():
    # Filtra o DataFrame para obter todas as informações do cliente com base no ID
    cliente = df_scaled[df_scaled['ID Cliente'] == cliente_id].iloc[0]

    # Chama a função de recomendação para o cliente atual
    recomendacao = sistema_recomendacao(cliente)

    # Verifica se há uma recomendação para o cliente atual antes de imprimir
    if recomendacao != "Sem recomendação neste momento":
        print(f"Cliente {cliente_id}: {recomendacao}")
```

## 2.2 Treinamento do Random Forest
