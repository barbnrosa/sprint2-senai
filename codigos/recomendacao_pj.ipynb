```python
# Função para recomendar produtos com base nas características do cliente
def recomendar_produtos(caracteristicas):
    filtered_df = df_pj.copy()
    for key, value in caracteristicas.items():
        filtered_df = filtered_df[filtered_df[key] == value]
    
    recommended_products = filtered_df['Produto GPOM'].unique()
    return recommended_products

# Exemplo de utilização da função de recomendação
caracteristicas_cliente = {'Área de Atuação': 'NÃO INFORMADO', 'Porte': 'PEQUENA', 'Código CNAE': '8640-2/01'}
recomendacoes = recomendar_produtos(caracteristicas_cliente)

print("Produtos Recomendados para o Cliente:")
for produto in recomendacoes:
    print(produto)
```
