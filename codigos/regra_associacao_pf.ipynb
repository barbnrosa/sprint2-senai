```python
import pandas as pd
```

```python
df_pf = pd.read_csv('../dados/PF_tratados.csv')
df_pf.head()
```

```python
transacao = []

for cliente_id in df_pf['ID Cliente'].unique():
    # Obter todas as transações do cliente
    transacoes_cliente = df_pf[df_pf['ID Cliente'] == cliente_id]
    
    # Agregar transações por produto e data de nascimento
    transacoes_agrupadas = transacoes_cliente.groupby(['Produto GPOM', 'Data de Nascimento']).size().reset_index(name='count')
    
    # Criar uma lista de produtos únicos por cliente
    lista_produtos = list(transacoes_agrupadas['Produto GPOM'])
    
    # Adicionar lista de produtos únicos à lista de transações
    transacao.append(lista_produtos)
```

## Algoritmo Apriori

### Aplicando a regra de associação

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder() # estanciar função
transacao_te = te.fit(transacao).transform(transacao)
```

```python
transacao_te
```

```python
# transformar em dataframe
transacao_transformado = pd.DataFrame(transacao_te, columns=te.columns_)
transacao_transformado
```

```python
# algoritmo apriori
from mlxtend.frequent_patterns import apriori
```

```python
items_frequentes_apriori = apriori(transacao_transformado, use_colnames=True, min_support=0.02)
items_frequentes_apriori.sort_values(['support'],ascending=False)
```

```python
from mlxtend.frequent_patterns import association_rules

regras_apriori = association_rules(items_frequentes_apriori, metric='confidence', min_threshold=0.2)
regras_apriori
```

```python
def recommend_items(customer_items, rules, min_confidence=0.6):
    """
    Função para recomendar itens com base nos itens comprados por um cliente
    e nas regras de associação geradas pelo algoritmo Apriori.
    """
    recommended_items = set()
    
    for _, regras_apriori in rules.iterrows():
        if set(regras_apriori['antecedents']).issubset(customer_items) and regras_apriori['confidence'] >= min_confidence:
            recommended_items.update(regras_apriori['consequents'])
    
    return recommended_items
```

```python
itens_comprados = {'EDUCACAO CONTINUADA'}
recommend_items = recommend_items(itens_comprados, regras_apriori)
print("Tens recomendados:", recommend_items)
```
